2. Your bar chart should have the names of the communes listed vertically along the y-axis and the x-axis should show the populations.

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
%matplotlib inline
import numpy as np

# Prepare the data
most_populated_communes = df.nlargest(10, 'Residents')

fig = plt.figure()
ax = fig.add_subplot(111)

bars = ax.barh(most_populated_communes['Commune'], most_populated_communes['Residents'])
# To display the value per bar
ax.bar_label(bars)
# To format the x-axis to show the population in thousands.
scale_x = 1e3
ticks_x = ticker.FuncFormatter(lambda x, pos: '{0:g}'.format(x/scale_x))
ax.xaxis.set_major_formatter(ticks_x)
# labels and title
ax.set_xlabel('Population (Thousand)')
ax.set_ylabel("Commune")
ax.set_title("Top 10 populated communes.")
plt.show()
```

# D. Probabilities
Compute the probability that a randomly selected commune with elevation over 2000 is from the canton of Valais.

I **assume** that canton Valais has the abbervation of VS as in the dataset cantons are represented with abberavations.
    


```python
import scipy.stats as stats
# Conditional Random sample of data 
sample = df[pd.to_numeric(df.Elevation)>2000].sample(1)
sample
```

```python
sample = df[pd.to_numeric(df.Elevation)>2000].sample(1)
sample
```

```python
# select rows that satisfy the condition of Elevation > 2000
df_high_elevation = df[df['Elevation'] > 2000]
from_vs = df_high_elevation[df_high_elevation['Canton'] == 'VS']
print(f'There are {len(df_high_elevation)} communes with Elevation of above 2000, which {len(from_vs)} are from canton VS.')
print(f'More than half the samples with high elevation are from canton VS. Therefore, I expect the probability to be around 0.5.')

```

```python
experiments = np.arange(100, 1100, 100)

for experiment in experiments:
    counter = 0
    for i in range(experiment):
        sample = df_high_elevation.sample(1)
        if sample['Canton'].item() == 'VS':
            counter += 1

    print(f'Iteration: {experiment} \t Probability: {np.round((counter / experiment), 3)}' )
```
